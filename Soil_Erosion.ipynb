```python
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import glob

# Step 3: Define Dataset Path (Modify based on your folder structure)
dataset_path = "/content/drive/MyDrive/archive_landslide/dataset"  # Change this to your actual dataset location
image_folder = os.path.join(dataset_path, "train/images/")   # Modify if needed
mask_folder = os.path.join(dataset_path, "train/masks/")   # Modify if needed
#usual train=80% || 70%

# Step 4: Load Images & Masks
def load_images_from_folder(folder, size=(256, 256), color_mode="rgb"):
    images = []
    for filename in sorted(glob.glob(folder + "/*.png")):  # Change extension if needed
        img = load_img(filename, target_size=size, color_mode=color_mode)
        img = img_to_array(img) / 255.0  # Normalize pixel values
        images.append(img)
    return np.array(images)

X_train = load_images_from_folder(image_folder)
Y_train = load_images_from_folder(mask_folder, size=(256, 256), color_mode="grayscale")  # Mask should be grayscale

print(f"✅ Loaded {len(X_train)} images and {len(Y_train)} masks.")

# Step 5: Define U-Net Model
def unet_model(input_size=(256, 256, 3)):
    inputs = Input(input_size)

    # Encoder
    c1 = Conv2D(64, (3, 3), activation="relu", padding="same")(inputs)
    c1 = Conv2D(64, (3, 3), activation="relu", padding="same")(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(128, (3, 3), activation="relu", padding="same")(p1)
    c2 = Conv2D(128, (3, 3), activation="relu", padding="same")(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    # Bottleneck
    c3 = Conv2D(256, (3, 3), activation="relu", padding="same")(p2)
    c3 = Conv2D(256, (3, 3), activation="relu", padding="same")(c3)

    # Decoder
    u4 = UpSampling2D((2, 2))(c3)
    m4 = Concatenate()([u4, c2])
    c4 = Conv2D(128, (3, 3), activation="relu", padding="same")(m4)
    c4 = Conv2D(128, (3, 3), activation="relu", padding="same")(c4)

    u5 = UpSampling2D((2, 2))(c4)
    m5 = Concatenate()([u5, c1])
    c5 = Conv2D(64, (3, 3), activation="relu", padding="same")(m5)
    c5 = Conv2D(64, (3, 3), activation="relu", padding="same")(c5)

    outputs = Conv2D(1, (1, 1), activation="sigmoid")(c5)

    model = Model(inputs, outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss="binary_crossentropy", metrics=["accuracy"])

    return model

model = unet_model()
model.summary()

# Step 6: Train the Model
history = model.fit(X_train, Y_train, epochs=10, batch_size=8, validation_split=0.2)
model_save_path = "/content/drive/MyDrive/unet_model.h5"
model.save(model_save_path)

# Step 7: Evaluate Model Performance
loss, accuracy = model.evaluate(X_train, Y_train)
print(f"✅ Model Accuracy: {accuracy:.2f}")

# Step 8: Visualize Predictions
def show_predictions(model, images, masks, num=3):
    preds = model.predict(images[:num])
    fig, axes = plt.subplots(num, 3, figsize=(10, num * 3))

    for i in range(num):
        axes[i, 0].imshow(images[i])
        axes[i, 0].set_title("Input Image")
        axes[i, 1].imshow(masks[i].squeeze(), cmap="gray")
        axes[i, 1].set_title("Ground Truth Mask")
        axes[i, 2].imshow(preds[i].squeeze(), cmap="gray")
        axes[i, 2].set_title("Predicted Mask")

    plt.tight_layout()
    plt.show()

show_predictions(model, X_train, Y_train)
```

```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 🔷 Set your dataset folder path (Update this!)
image_folder = '/content/drive/MyDrive/HerbaceousVegetation'  # 🔄 Change this to your dataset path

# 🔷 Parameters
image_size = (128, 128)  # Resize images to save memory
n_clusters = 5  # Adjust based on soil type variations
batch_size = 5000  # For MiniBatchKMeans to handle large datasets
num_images_to_load = 20  # Load limited images for testing (Adjust if needed)

# 🔷 Load and preprocess images
images = []

for i, image_name in enumerate(os.listdir(image_folder)):
    if i >= num_images_to_load:
        break  # Stop after loading the specified number of images
    image_path = os.path.join(image_folder, image_name)
    if image_path.endswith('.jpg') or image_path.endswith('.png'):
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
        image_resized = cv2.resize(image, image_size)  # Resize image
        images.append(image_resized)

# Convert to numpy array
images = np.array(images)

# 🔷 Reshape images to (num_pixels, 3) for clustering
pixels = images.reshape(-1, 3)  # Flatten pixels (R, G, B) for clustering

# 🔷 Standardize pixel values for better clustering performance
scaler = StandardScaler()
pixels_scaled = scaler.fit_transform(pixels)

# 🔷 Apply PCA for dimensionality reduction (Optional for large datasets)
pca = PCA(n_components=2)  # Reduce features to 2 dimensions
pixels_pca = pca.fit_transform(pixels_scaled)

# 🔷 Use MiniBatchKMeans to cluster pixels efficiently
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=batch_size, random_state=42)
kmeans.fit(pixels_pca)  # Fit model on reduced data

# Get the cluster labels (assign a soil type to each pixel)
labels = kmeans.labels_

# 🔷 Reshape labels back to match the image dimensions
labels_reshaped = labels.reshape(len(images), image_size[0], image_size[1])

# 🔷 Visualize the clustered soil types for the first image
plt.figure(figsize=(8, 6))
plt.imshow(labels_reshaped[0], cmap='tab10')  # Show clusters for the first image
plt.title('Clustered Soil Types (First Image)')
plt.axis('off')
plt.show()

# 🔷 Show cluster centers (average color representation of soil types)
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
plt.figure(figsize=(8, 2))
plt.imshow([cluster_centers.astype(int)])  # Display cluster colors
plt.title('Cluster Centers (Soil Types)')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

# 📌 Load and merge CSVs (Drop first column)
file_paths = ["/content/drive/MyDrive/2022_res.csv", "/content/drive/MyDrive/2023_res.csv", "/content/drive/MyDrive/2024_res.csv"]
dfs = [pd.read_csv(file).iloc[:, 1:] for file in file_paths]  # Drop first column
df = pd.concat(dfs, axis=0, ignore_index=True)  # Merge all years

# 📌 Define Features & Target
features = ['latitude', 'longitude', 'Annual_Rainfall', 'Slope_Aspect',
            'Slope_Gradient', 'Slope_Length', 'C_Factor', 'P_Factor',
            'K', 'R_Factor', 'LS_Factor']
target = 'Soil_Erosion'

# 📌 Shuffle data to prevent bias
df = shuffle(df, random_state=42)

# 📌 Check Soil_Erosion distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Soil_Erosion'], bins=20, kde=True)
plt.title("Soil Erosion Distribution")
plt.show()

# 📌 StandardScaler for normalization
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])

# 📌 Split Data
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# 📌 Train Improved Random Forest Model
model = RandomForestRegressor(
    n_estimators=300,         # More trees for better learning
    max_depth=30,             # Deeper trees to capture complex patterns
    min_samples_split=5,      # Prevent overfitting
    random_state=42
)
model.fit(X_train, y_train)
import joblib

# Save the model
joblib.dump(model, '/content/drive/MyDrive/random_forest_model.pkl')
# 📌 Feature Importance Analysis
importances = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
print("\nFeature Importance:\n", feature_importance.sort_values(by='Importance', ascending=False))

# 📌 Function to Predict Soil Erosion for Specific Location
def predict_soil_erosion(latitude, longitude):
    location_data = df[(df['latitude'] == latitude) & (df['longitude'] == longitude)]

    if location_data.empty:
        print("⚠️ No exact match found. Using closest available data.")
        location_data = df.iloc[:5]  # Use closest available data

    prediction = model.predict(location_data[features])

    # Compare with last year's erosion
    last_year_erosion = location_data[target].mean()
    trend = "INCREASE" if np.mean(prediction) > last_year_erosion else "DECREASE"

    print(f"\n📍 Predicted Soil Erosion for Next Year at ({latitude}, {longitude}): {np.mean(prediction):.3f}")
    print(f"📊 Trend: {trend}")

# 📌 Example Prediction
predict_soil_erosion(23.0714, 75.957)
```
